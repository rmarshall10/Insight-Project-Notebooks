# Loading image bounding boxes

Images from OpenImage 4

```python
import pandas as pd
import numpy as np
```

```python
DIRECTORY = ""
VAL_PATH = "OID/csv_folder/validation-annotations-bbox.csv"
TEST_PATH = "OID/csv_folder/test-annotations-bbox.csv"
football = '/m/01226z'
```

```python
iter_csv = pd.read_csv(DIRECTORY + VAL_PATH, iterator=True, chunksize=1000)
df_val = pd.concat([chunk[chunk['LabelName'] == football] for chunk in iter_csv])
```

```python
iter_csv2 = pd.read_csv(DIRECTORY + TEST_PATH, iterator=True, chunksize=1000)
df_test = pd.concat([chunk[chunk['LabelName'] == football] for chunk in iter_csv2])
```

```python
df = pd.concat([df_val, df_test])
```

```python
df.head()
```

The coordinates are from the top left corner, and are %of image, not pixels, etc.

```python
df2 = df[['ImageID','XMin','XMax','YMin','YMax']]
```

```python
df2.head()
```

```python
df2.reset_index(drop=True)
```

```python
df2['XMin'] = df2['XMin'].apply(lambda x: int(300*x))
df2['XMax'] = df2['XMax'].apply(lambda x: int(300*x))
df2['YMin'] = df2['YMin'].apply(lambda x: int(300*x))
df2['YMax'] = df2['YMax'].apply(lambda x: int(300*x))
```

```python
df2.head()
```

```python
df2['width'] = 300
df2['height'] = 300
df2['class'] = 'soccer_ball'
```

```python
df2.head()
```

```python
df2 = df2.rename(columns={"ImageID":"filename", "XMin":"xmin", "XMax":"xmax","YMin":"ymin","YMax":"ymax"})
```

```python
df2.head()
```

```python
cols = ["filename","width","height","class","xmin","ymin","xmax","ymax"]
df3 = df2[cols]
```

```python
df3.head()
```

```python
df3['filename'] = df3['filename'].apply(lambda x: x + ".jpg")
```

```python
df3.head()
```

```python
df3.to_csv('OI_boxes_test_and_val.csv', index=False)
```
